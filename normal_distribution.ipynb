```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy.stats import shapiro
```

```python
# Load the data
df = pd.read_csv('./historicalData/processed_btc_data.csv')
```

```python
# Plot the distribution of daily returns
plt.figure(figsize=(10, 6))
plt.hist(df['log_returns'], bins=50, edgecolor='black', alpha=0.7)
plt.title('Distribution of Daily Log Returns of BTC')
plt.xlabel('Log Return')
plt.ylabel('Frequency')
plt.show()
```

```python
# Q-Q plot to check normality
plt.figure(figsize=(8, 8))
stats.probplot(df['log_returns'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Log Returns')
plt.show()
```

```python
# Shapiro-Wilk Test
from scipy.stats import shapiro
log_returns = df['log_returns']
# Perform Shapiro-Wilk test for normality
stat, p_value = shapiro(log_returns)
# Print the results
print(f'Shapiro-Wilk Test Statistic: {stat}')
print(f'P-value: {p_value}')
# Interpretation
if p_value > 0.01:
    print("The data is likely to follow a normal distribution (fail to reject H0).")
else:
    print("The data does not follow a normal distribution (reject H0).")
```
